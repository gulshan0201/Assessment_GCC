Given a number by the user, write a program to print the Octal and Hexadecimal values of that number as result.

```python
# Read input
num = int(input())

# Print Decimal
print(num)

# Print Octal (without 0o prefix)
print(oct(num)[2:])

# Print Hexadecimal in lowercase (without 0x prefix)
print(hex(num)[2:])
```

Write a program to read a 4-digit number (If the number is more than 4-digits then ignore whatever given after 4-digits) and find the sum of the first and last digit of this number.

```python
num = int(input())

num = num // (10 ** (len(str(num)) - 4)) if len(str(num)) > 4 else num

first_digit = num // 1000
last_digit = num % 10

print(first_digit + last_digit)
```
